## Graph class to represent the graph and generate data recrusivly 

```python
# Python program to print all paths from a source to destination.
  
from collections import defaultdict
import time
import random

# This class represents a directed graph
# using adjacency list representation
class Graph:
  
    def __init__(self, data):
        # No. of vertices
        self.V = len(data)+1
         
        # default dictionary to store graph
        self.graph = data
        
  
    '''A recursive function to print all paths from 'u' to 'd'.
    visited[] keeps track of vertices in current path.
    path[] stores actual vertices and path_index is current
    index in path[]'''
    def getAllPathsUtil(self, u, d, l, visited, path, n, min_sentences, max_sentences):
        if len(self.allPath) < n:

            # Mark the current node as visited and store in path
            visited[u]+=1
            path.append(l)

            # If current vertex is same as destination, then print
            # current path[]
            if u == d:
                if max_sentences > len(path) > min_sentences:
                    self.allPath.append(path.copy())
            else:
                # If current vertex is not destination
                # Recur for all the vertices adjacent to this vertex
    #             vn = random.choice([1,2,3])
                random.shuffle(self.graph[u])
                for i in self.graph[u]:
                    if visited[i[-1]]< 1:
                        self.getAllPathsUtil(i[-1], d, i, visited, path, n, min_sentences, max_sentences)

            # Remove current vertex from path[] and mark it as unvisited
            path.pop()
            visited[u]= 0
  
  
    # Prints all paths from 's' to 'd'
    def getAllPaths(self, s, d, n=100,min_sentences=10, max_sentences=20):
        tic = time.time()
 
        # Mark all the vertices as not visited
        visited =[0]*(self.V)
        
 
        # Create an array to store paths
        path = []
        self.allPath = []
 
        # Call the recursive helper function to print all paths
        self.getAllPathsUtil(s, d, [s], visited, path, n, min_sentences, max_sentences)
        allPath = self.allPath
        self.allPath = []
        toc = time.time()
        print('create data time (s): ', toc-tic)
        return allPath
    
    def create_text_pair_template(self, path):
        sent = {'parms':defaultdict(list),'text':''}
        pcnt = defaultdict(int)
        osent = {'parms':defaultdict(list),'text':''}
        opcnt = defaultdict(int)
        last_text = []
        new_path = []
        for cc in path:
            
            ccc = cc.copy()
            con_num = 2
            if len(cc) >= 3:
                if '3 alternative' not in last_text[-2]:
                    ccc.remove(cc[-2])
                    con_num = 3
            
                    
            new_path.append(ccc)
#             print(last_text,ccc)
            for c in ccc:
                k = 'C{}'.format(str(c))
                t=random.choice(range(len(st_templates['St'][k])))
                s = random.choice(range(len(st_templates['St'][k][t])))
                s2 = random.choice(range(len(st_templates['St2'][k][t])))

                sent['text'] += st_templates['St'][k][t][s]['template']+ ' '

                for p in st_templates['St'][k][t][s]['parms']:
                    if p=='source':
                        tmp = len(st_templates['St'][k][t][s]['parms'][p])
                        if tmp > 0:
                            for ii,pp in enumerate(sent['parms']['target'][-tmp:]):
                                sent['parms'][p].append(pp)
                                sent['text'] = sent['text'].replace(st_templates['St'][k][t][s]['parms'][p][ii-1],pp)
                                
                    elif p =='target':
                        for pp in st_templates['St'][k][t][s]['parms'][p]:
                            tmp_t = '<T_{}>'.format(str(pcnt[p]))
                            sent['parms'][p].append(tmp_t)
                            sent['text'] = sent['text'].replace(pp,tmp_t)
                            pcnt[p]+=1
                            
                    elif p =='event':
                        for pp in st_templates['St'][k][t][s]['parms'][p]:
                            tmp_t = '<{}_{}>'.format(pp.split('-')[0],str(pcnt[p]))
                            sent['parms'][p].append(tmp_t)
                            sent['text'] = sent['text'].replace(pp,tmp_t)
                            pcnt[p]+=1
#                     elif p == 'condithion' and cc in [[18], [19], [20], [21], [22], [23], [24], [25], [26]]:
# #                         tmp = len(st_templates['St'][k][t][s]['parms'][p])
# #                         if tmp > 0:
# #                             for ii,pp in enumerate():
#                         pp = sent['parms'][p][-con_num]
#                         sent['parms'][p].append(pp)
#                         sent['text'] = sent['text'].replace(st_templates['St'][k][t][s]['parms'][p][0],pp)
        
                    else:
                        for pp in st_templates['St'][k][t][s]['parms'][p]:
                            tmp_t = '<{}_{}>'.format(p,str(pcnt[p]))
                            sent['parms'][p].append(tmp_t)
                            sent['text'] = sent['text'].replace(pp,tmp_t)
                            pcnt[p]+=1

                osent['text'] += st_templates['St2'][k][t][s2]['template']+ ' '
                last_text.append(st_templates['St2'][k][t][s2]['template'])

                for p in st_templates['St2'][k][t][s2]['parms']:
                    if p=='source':
                        tmp = len(st_templates['St2'][k][t][s2]['parms'][p])
                        if tmp > 0:
                            for ii,pp in enumerate(sent['parms']['target'][-tmp:]):
                                osent['parms'][p].append(pp)
                                osent['text']= osent['text'].replace(st_templates['St2'][k][t][s2]['parms'][p][ii-1],pp)
                                
                    elif p =='target':
                        for pp in st_templates['St2'][k][t][s2]['parms'][p]:
                            tmp_t = '<T_{}>'.format(str(opcnt[p]))
                            osent['parms'][p].append(tmp_t)
                            osent['text']= osent['text'].replace(pp,tmp_t)
                            opcnt[p]+=1
                            
                    elif p =='event':
                        for pp in st_templates['St2'][k][t][s2]['parms'][p]:
                            tmp_t = '<{}_{}>'.format(pp.split('-')[0],str(opcnt[p]))
                            osent['parms'][p].append(tmp_t)
                            osent['text'] = osent['text'].replace(pp,tmp_t)
                            opcnt[p]+=1
                    else:
                        for pp in st_templates['St2'][k][t][s2]['parms'][p]:
                            tmp_t = '<{}_{}>'.format(p,str(opcnt[p]))
                            osent['parms'][p].append(tmp_t)
                            osent['text'] = osent['text'].replace(pp,tmp_t)
                            opcnt[p]+=1
        return [new_path,sent, osent]

            
        
    def create_data(self,s=[14,15],d=[2,7,13,28],n=10000,min_sentences=10, max_sentences=20):
        paths = []
        tmp_n = n // (len(s)*len(d))
        for ss in s:
            for dd in d:
                paths.extend(self.getAllPaths(ss, dd ,n=tmp_n, min_sentences=min_sentences, max_sentences=max_sentences))
        random.shuffle(paths)
        paths = paths[:n]
        print ("Number of different paths from % d :" %(len(paths)))
        
        sentneces = []
        for path in paths:
            sentneces.append(self.create_text_pair_template(path))
        
        return sentneces


#  Create a graph given in the above diagram
g = Graph(graph_templates)
paths = g.create_data(n=1000000,min_sentences=4, max_sentences=20)
paths[0]
```

## add real BPMN elements to generated examples

```python

def create_data_with_bpmn_elements(paths, bpmn_elements):
    data = []
    for i, path in enumerate(paths):        
        if len(path[1]['parms']['target']) == len(path[2]['parms']['target']):
            parms = {}
            for k in path[1]['parms']:
                for elem in path[1]['parms'][k]:
                    if elem not in parms:
                        if k == 'target':
                            new_elem = random.choice(bpmn_elements['activities'])
                        elif k == 'source':
                            pass
                        elif k == 'role':
                            new_elem = random.choice(bpmn_elements['roles'])
                        elif k == 'object':
                            new_elem = random.choice(bpmn_elements['onjects'])
                        elif k == 'condition':
                            pass
                        elif k == 'event':
                            new_elem = random.choice(bpmn_elements[elem.split('_')[0].replace('<','')])
                            
                        else:
                            new_elem = random.choice(bpmn_elements[k])
                        parms[elem] = new_elem.replace('\n','')
#             print(i,path[1])
            process_text = path[1]['text']
            sound_process_text = path[2]['text']
            for k in parms:
                process_text = process_text.replace(k,parms[k])
                sound_process_text = sound_process_text.replace(k,parms[k])
            data.append({"process_text":process_text,"sound_process_text":sound_process_text})
                
    return data

create_data_with_bpmn_elements(paths, bpmn_elements) 
```
